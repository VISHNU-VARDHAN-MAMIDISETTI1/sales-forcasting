**Feature Engineering Create New Columns**

```python
df4["Month"]=df4["InvoiceDate"].dt.month
df4.head()
```

### Visualization & EDA

#### Plot Monthly Sales

```python
monthly_sales=df4.groupby("Month")["TotalPrice"].sum()
monthly_sales.plot(kind="line",title="Monthly Sales",marker="o")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.grid()
plt.show()
```

**Insights**


Total sales started rising up in August having a peek in November.This is likely due to holiday season at the end of the year

**Top 5 Countries**

```python
top_5_countries=df4.groupby("Country")['TotalPrice'].sum().sort_values(ascending=False).head(5)
top_5_countries.plot(kind="barh",title="Top 5 Countries by Sales")
plt.xlabel("Total Sales")
plt.ylabel("Country")
plt.grid(axis='x')
plt.show()
```

```python
country_wise_sales=df4.groupby("Country")["TotalPrice"].sum()
total_sales=country_wise_sales.sum()

top_5_countries=df4.groupby("Country")['TotalPrice'].sum().sort_values(ascending=False).head(5)
percentage_sales=(top_5_countries/total_sales)*100
plt.figure(figsize=(10,6))
bars=plt.barh(top_5_countries.index,percentage_sales)
plt.xlabel("Percentage of Total Sales")
plt.ylabel("Country")
plt.title("Top 5 Countries by Sales Percentage")
plt.grid(axis='x')
for bar, percentage in zip(bars,percentage_sales):
  plt.text(bar.get_width()+1,bar.get_y()+bar.get_height()/2,f"{percentage:.2f}%",ha="left",va="center")
plt.show()
```

```python
product_sales=df4.groupby('StockCode')['TotalPrice'].sum()
top_5_products=product_sales.sort_values(ascending=False).head(5)
top_5_products.plot(kind="barh",color="red",title="Top 5 Products by Sales")
plt.xlabel("Total Sales")
plt.ylabel("Product")
plt.gca().invert_yaxis()
plt.grid(axis='x')
plt.show()
```

# RFM Analysis

```python
current_dt=df4["InvoiceDate"].max()+ pd.Timedelta(days=1)
current_dt
```

```python
rfm=df4.groupby("CustomerID").agg({
    "InvoiceDate":lambda x:(current_dt-x.max()).days,
    "InvoiceNo":"count",
    "TotalPrice":"sum"
})
rfm.columns=["Recency","Frequency","Monetary"]
rfm.head()
```

```python
df4[df4.CustomerID==12346]
```

# Segment Customer Based on RFM

```python
rfm['R_Segment']=pd.qcut(rfm['Recency'],4,labels=[4,3,2,1])
rfm['F_Segment']=pd.qcut(rfm["Frequency"],4,labels=[1,2,3,4])
rfm['M_Segment']=pd.qcut(rfm["Monetary"],4,labels=[1,2,3,4])
rfm["RFM_Score"]=rfm[['R_Segment','F_Segment','M_Segment']].sum(axis=1)
rfm
```

```python
rfm.sort_values("RFM_Score",ascending=False)
```

```python
customer_last_purchase=df4.groupby("CustomerID")['InvoiceDate'].max()
customer_last_purchase.head(5)
```

```python
customer_last_purchase=(current_dt-customer_last_purchase).dt.days
customer_last_purchase.head(5)
```

**Churn Threshold**

```python
churn_threshold=90
churn_customer=customer_last_purchase[customer_last_purchase>churn_threshold]
churn_customer.head()
```

```python
print("Number of Churned Customers:",len(churn_customer))
plt.figure(figsize=(10,6))
plt.hist(customer_last_purchase,bins=70,color="green",alpha=0.7)
plt.axvline(x=churn_threshold,color="red",linestyle="--",label="Churn Threshold")
plt.xlabel("Days Since Last Purchase")
plt.ylabel("Number of Customers")
plt.title("Distribution of Days Since Last Purchase")
plt.legend()
```
